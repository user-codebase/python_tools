## W Pythonie funkcje tworzymy za pomocą słowa kluczowego def, po którym następuje nazwa funkcji, a następnie nawiasy okrągłe, w których mogą znaleźć się parametry. Po dwukropku rozpoczynamy blok kodu, który będzie częścią funkcji

```python
def dodaj(a, b):
    return a + b
```

## Aby wywołać funkcję, używamy jej nazwy, a następnie nawiasów okrągłych, w których przekazujemy argumenty, odpowiadające parametrom funkcji.

```python
dodaj(1, 2)
```
