```python
salaries = [1000, 5000, 7000]
salary_iterator = iter(salaries)
print(next(salary_iterator))
print(next(salary_iterator))
print(next(salary_iterator))
```

```python
salary_iterator = iter(salaries)
salary_sum = 0
for salary in salary_iterator:
	salary_sum = salary_sum + salary
print("Total salaries = %d" % salary_sum)
```
